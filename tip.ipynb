```python
# Importing Important Liberaries

import numpy as np
import pandas as pd
from pandas import Series, DataFrame
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
```

```python
df = pd.read_csv('C:\\Users\\Gagandeep\\Downloads\\tips.csv')
df
```

# 1. Five Number Summary For bill and tip

```python
# Five NO. Summary for Bill Column
bill = df.total_bill

print('Minimum Bill:',np.min(bill))
print('Maximum Bill:',np.max(bill))
print('Standard Deviation of Bill:',np.std(bill))
print('Mean Bill:',np.mean(bill))
print('Median Bill:',np.median(bill))
```

```python
# Five NO. Summary for tip Column
tip = df.tip

print('Minimum Tip:',np.min(tip))
print('Maximum Tip:',np.max(tip))
print('Standard Deviation of Tip:',np.std(tip))
print('Mean Tip:',np.mean(tip))
print('Median Tip:',np.median(tip))
```

# 5. Find out the IQR

```python
# IQR Value
bill_tip = pd.DataFrame(df,columns=['total_bill','tip'])
print("IQR For Total Bill : ",stats.iqr(bill))
print("IQR For Tip : ",stats.iqr(tip))
print("IQR For Total Bill & Tip : ",stats.iqr(bill_tip))
```

# 7. Plot the cumulative Frequency Polygon for "Total Bill" and "Tip"

```python
# Plotting Cumulative Frequency Polygon for "Total Bill" Column

values, base = np.histogram(bill, bins=40)
cumulative = np.cumsum(values)
plt.plot(base[:-1], cumulative)
plt.title('Total Bill',fontsize=15)
plt.grid(axis='y', alpha=0.75)
plt.grid(axis='x', alpha=0.75)
plt.show()
```

```python
# Plotting Cumulative Frequency Polygon for "Tip" Column

values, base = np.histogram(tip, bins=40)
cumulative = np.cumsum(values)
plt.plot(base[:-1], cumulative)
plt.title('Tip',fontsize=15)
plt.grid(axis='y', alpha=0.75)
plt.grid(axis='x', alpha=0.75)
plt.show()
```

# 8. Prepare the Frequency Table and Bar Chart for "Size". Summarize and explain your findings

```python
# Frequency table and Bar Chart

print("Frequency Table for Size")
size = pd.crosstab(index=df['size'],columns='count')
print(size)
size.plot.bar(figsize=(15,5),rot=0)
```

```python
# A little summary of above chart
print("At first we Created a frequency table for size column in dataset using crosstab() function. Then using the frequency table we created Bar-Chart."+'\n'+"By creating the table and graph we found out the numbers of the families with different sizes (families containing different numbers of peoples) that visited our Resturant!!!!")
```

# 9. Prepare two variable frequency tables for "Size" vs "Tip" and "Size" vs "Total Bill". Summarize and explain your findings.

```python
#Size Vs Tip

size_tip = pd.crosstab(index=df["size"],columns=df['tip'], margins=True)
size_tip.index=["1 member","2 members","3 members","4 members","5 members","6 members","Total"]
print(size_tip)
print("\n Through this table we get to know that how much tip we recieve from families with different number of peoples.")
```

```python
#Size Vs Total Bill

size_bill = pd.crosstab(index=df['size'],columns=df['total_bill'],margins=True)
size_bill.index=["1 member","2 members","3 members","4 members","5 members","6 members","Total"]
print(size_bill)
print("\n Through this table we get to know that how much worth of food did families buy families with different number of peoples.")
```
